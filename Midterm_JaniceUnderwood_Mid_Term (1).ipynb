```python
# 06-06 Encoding: Converting categorical data to numbers
#Several techniques can be used, depending on the nature of the categorical variable (nominal or ordinal) and the requirements of the machine learning model
#One-Hot Encoding / Label Encoding (Ordinal Encoding) / Custom Mapping / pd.factorize()
#Encoding Categorical Variables
# One-hot encode categorical columns
#Separate Target Variable
# X is the import data / independent variable and Y is the dependable variable / target variable

target_col = 'income'
X = df.drop(columns=[target_col])
y = df[target_col]

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# 06-06 Encoding: Converting categorical data to numbers
#Several techniques can be used, depending on the nature of the categorical variable (nominal or ordinal) and the requirements of the machine learning model
#One-Hot Encoding / Label Encoding (Ordinal Encoding) / Custom Mapping / pd.factorize()
#Encoding Categorical Variables
# One-hot encode categorical columns
#Feature Categorization
# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Categorical Columns:", categorical_cols)
print("\nNumerical Columns:", numerical_cols)
```

```python
# 06-06 Encoding: Converting categorical data to numbers
#Several techniques can be used, depending on the nature of the categorical variable (nominal or ordinal) and the requirements of the machine learning model
#One-Hot Encoding / Label Encoding (Ordinal Encoding) / Custom Mapping / pd.factorize()
#Encoding Categorical Variables
# One-hot encode categorical columns
encoder = OneHotEncoder(drop='first', sparse_output=False)

encoded_cols = pd.DataFrame(
    encoder.fit_transform(X[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols)
)

# Merge encoded columns with numerical features
X_encoded = pd.concat([X[numerical_cols], encoded_cols], axis=1)

print("Shape after encoding:", X_encoded.shape)
X_encoded.head()
```

```python
# Project : Predict whether a person will be making moreÂ thanÂ $50k
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# One-hot encode the categorical columns in df_processed
categorical_cols = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country', 'income']
df_processed_encoded = pd.get_dummies(df_processed, columns=categorical_cols, drop_first=True)

TARGET_COL = 'income_>50K'

X = df_processed_encoded.drop(columns=[TARGET_COL])
y = df_processed_encoded[TARGET_COL]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

model = LogisticRegression(solver='liblinear', random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]


print("\n--- Logistic Regression Model Evaluation ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
print(f"AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")

# The code below provides a solution to predict income bracket for making more than 50k using the adult.csv dataset,
# covering all necessary steps from data cleaning and preprocessing to model training and evaluation using LogisticÂ Regression.
# Accurancy: The overall percentage of predictions the model got
# Fi-Score: providing a balanced measure of performance.
# AUC means evaluation.The model has a high ability to distinguish between the two incomeÂ classes.
```
